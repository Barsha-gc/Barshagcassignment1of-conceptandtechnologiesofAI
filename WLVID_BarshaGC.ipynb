```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(
    "Human_Development_Index_Dataset.csv",
    encoding="latin1"
)

df.head()
```

```python
# Remove duplicate rows
hdi_2022_df = hdi_2022_df.drop_duplicates()

# Convert selected columns to numeric
numeric_cols = ['hdi', 'life_expectancy', 'gross_inc_percap']

for col in numeric_cols:
    hdi_2022_df[col] = pd.to_numeric(
        hdi_2022_df[col],
        errors='coerce'
    )
```

```python
high_hdi = hdi_2022_df[hdi_2022_df['hdi'] > 0.8]
high_hdi.sort_values('gross_inc_percap', ascending=False).head(10)
```

```python
def hdi_category(value):
    if value < 0.55:
        return "Low"
    elif value < 0.70:
        return "Medium"
    elif value < 0.80:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(hdi_category)
hdi_2022_df.head()
```

```python
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

```python
df_2020_22 = df[df['year'].isin([2020, 2021, 2022])].copy()
df_2020_22.to_csv("HDI_problem1B.csv", index=False)

df_2020_22.head()
```

```python
countries = ['Nepal', 'India', 'Bangladesh', 'Sri Lanka', 'Pakistan']

plt.figure(figsize=(8,5))
for c in countries:
    subset = df_2020_22[df_2020_22['country'] == c]
    plt.plot(subset['year'], subset['hdi'], marker='o', label=c)

plt.xlabel("Year")
plt.ylabel("HDI")
plt.title("HDI Trends (2020â€“2022)")
plt.legend()
plt.show()
```

```python
south_asia = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

sa_df = df[df['country'].isin(south_asia)].copy()
sa_df.to_csv("HDI_SouthAsia.csv", index=False)
sa_df.head()
```

```python
sa_df['Composite Score'] = (
    0.30 * sa_df['life_expectancy'] +
    0.30 * sa_df['gross_inc_percap']
)

sa_df.sort_values('Composite Score', ascending=False).head()
```

```python
middle_east = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

sa_20_22 = df_2020_22[df_2020_22['country'].isin(south_asia)]
me_20_22 = df_2020_22[df_2020_22['country'].isin(middle_east)]

sa_20_22.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
me_20_22.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)

sa_20_22.head(), me_20_22.head()
```

```python
## Regional Classification

Countries are manually grouped into **South Asia** and **Middle East** based on
geographical classification, as required by the assignment tasks.
Only these two regions are included in the regional analysis; countries outside
these regions are intentionally excluded.
```

```python
region_map = {
    # South Asia
    "Afghanistan": "South Asia",
    "Bangladesh": "South Asia",
    "Bhutan": "South Asia",
    "India": "South Asia",
    "Maldives": "South Asia",
    "Nepal": "South Asia",
    "Pakistan": "South Asia",
    "Sri Lanka": "South Asia",

    # Middle East
    "Bahrain": "Middle East",
    "Iran": "Middle East",
    "Iraq": "Middle East",
    "Israel": "Middle East",
    "Jordan": "Middle East",
    "Kuwait": "Middle East",
    "Lebanon": "Middle East",
    "Oman": "Middle East",
    "Palestine": "Middle East",
    "Qatar": "Middle East",
    "Saudi Arabia": "Middle East",
    "Syria": "Middle East",
    "United Arab Emirates": "Middle East",
    "Yemen": "Middle East"
}
```

```python
df['region'] = df['country'].map(region_map)
df_2020_22['region'] = df_2020_22['country'].map(region_map)
hdi_2022_df['region'] = hdi_2022_df['country'].map(region_map)
```

```python
Q1 = sa_df['hdi'].quantile(0.25)
Q3 = sa_df['hdi'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_hdi = sa_df[(sa_df['hdi'] < lower) | (sa_df['hdi'] > upper)]
outliers_hdi[['country', 'hdi']]
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    data=sa_df,
    x='gross_inc_percap',
    y='hdi',
    color='blue',
    label='Normal'
)

sns.scatterplot(
    data=outliers_hdi,
    x='gross_inc_percap',
    y='hdi',
    color='red',
    label='Outliers'
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Outlier Detection: South Asia")
plt.legend()
plt.show()
```

```python
sns.lmplot(
    data=sa_df,
    x='gender_development',
    y='hdi',
    height=5,
    aspect=1.2
)
plt.title("HDI vs Gender Development Index")
plt.show()
```

```python
sa_df['GNI_HDI_Gap'] = sa_df['gross_inc_percap'] - sa_df['hdi']

sa_df.sort_values('GNI_HDI_Gap', ascending=False).head(3)
```

```python
sa_df.sort_values('GNI_HDI_Gap').head(3)
```

```python
top_gap = sa_df.sort_values('GNI_HDI_Gap', ascending=False).head(3)
low_gap = sa_df.sort_values('GNI_HDI_Gap').head(3)

gap_df = pd.concat([top_gap, low_gap])

plt.figure(figsize=(8,5))
sns.barplot(
    data=gap_df,
    x='country',
    y='GNI_HDI_Gap'
)

plt.title("Top Positive and Negative GNIâ€“HDI Gaps")
plt.xticks(rotation=45)
plt.show()
```
